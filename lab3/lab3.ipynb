```python
import pandas as pd
import re
from scipy.spatial import distance
```

```python
name_dict = {
    'gra_o_tron': 'game_of_thrones',
    'miasteczko_south_park': 'south_park',
    'southpark': 'south_park',
    'bladerunner': 'blade_runner',
    'fantastyczne_zwierzęta_i_jak_je_znaleźć': 'fantastyczne_zwierzeta',
    'fightclub': 'fight_club',
    'ojciec_chrzesny_i': 'ojciec_chrzestny', # XD
    'gwiezdne_wojny_vii': 'star_wars_vii',
    'nienawistna_ósemka': 'hateful_8',
    'jestem_legendą': 'i_am_legend',
    'księgowy': 'the_accountant',
    'limitlessjestem_bogiem': 'limitless',
    'jestem_bogiem_nie_polski': 'limitless',
    'list_do_m': 'listy_do_m',
    'niezwykły_przypadek_brnjamina_buttona': 'ciekawy_przypadek_benjamina_buttona',
    'siedem': 'se7en',
    'zmerzch': 'zmierzch', # XDD
}

def normalize_names(name): 
    name = str(name)
    name = name.lower()
    name = re.sub(r'\s+', '_', name)
    name = re.sub(r'\W', '', name)
    name = re.sub(r'_+', '_', name)
    return name_dict[name] if name in name_dict else name
```

```python
# Read csv file
df = pd.read_csv('data.csv', sep=";")
df.dropna(axis='columns')
df
```

```python
name_movies_dict = {}
movies_list = set()
for row in df.iterrows():
    movies_dict = {}
    movies = row[0][1:]
    for i, movie in enumerate(movies):
      movie = normalize_names(movie)
      if i%2 != 0 or movie == 'nan' or i >= len(movies) - 1:
        continue
      movies_dict[movie] = movies[i+1] 
      movies_list.add(movie)
    name = (row[0][0])
    name_movies_dict[name] = movies_dict
name_movies_dict
```

```python
sorted_movies = sorted(list(movies_list))
```

```python
name_movies_matrix = {}
for name, user_movies in name_movies_dict.items():
    movies = []
    for movie in sorted_movies:
        if movie not in user_movies:
            movies.append(0)
        else:
            movies.append(user_movies[movie])
    name_movies_matrix[name] = movies
name_movies_matrix
```

```python
def calculate_distance(user1, user2):
    user_1_movies = (name_movies_matrix[user1])
    user_2_movies = (name_movies_matrix[user2])
    return distance.euclidean(user_1_movies, user_2_movies)

def get_name_distance_dict(user):
    name_distance_dict = {}
    for name in name_movies_matrix:
        if name == user:
            continue
        name_distance_dict[name] = calculate_distance(user, name)
    name_distance_dict = {k: v for k, v in sorted(name_distance_dict.items(), key=lambda item: item[1])}
    return name_distance_dict

def recommend_movies(user):
    name_distance_dict = get_name_distance_dict(user)
    user_movies = name_movies_dict[user]
    recommended_movies = []
    for name in name_distance_dict:
        for movie, score in name_movies_dict[name].items():
            if movie not in user_movies and movie not in recommended_movies and score > 5:
                recommended_movies.append(movie)
            if len(recommended_movies) == 5:
                return recommended_movies
    return recommended_movies
```

```python
recommend_movies("Jakub Jabłoński")
```
